```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, activation='softmax'),
    tf.keras.layers.Dense(32, activation='tanh'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(128, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid'), # 0-1 사이 output 추출 sigmoid function
])

# optimiaer list: adam, sgd, rmsprop, adagrad, adadelta, adamax, nadam
# loss list: binary_crossentropy, categorical_crossentropy, sparse_categorical_crossentropy, mean
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# epochs: 전체 데이터셋이 모델을 통과하는 횟수
# batch_size: 모델이 가중치를 업데이트하기 전에 처리하는 샘플 수
# x_train, y_train 데이터 정의하기
# model.fit(x_train, y_train, epochs=10, batch_size=32)
```
